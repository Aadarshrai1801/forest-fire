### Algerian Forest Fire Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("ForestFire.csv", header = 1)
```

```python
df.head()
```

```python
df.info()
```

### Data Cleaning

```python
# Missing values
df[df.isnull().any(axis = 1)]
```

```python
df.loc[:122, "Region"] = 0
df.loc[122:, "Region"] = 1
```

```python
df.info()
```

```python
df.head()
```

```python
df[["Region"]].isnull().sum()
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.isnull().sum()
```

```python
# Removing Null values
df = df.dropna().reset_index(drop = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
# Change the required columns to int data type
df[["month", "day", "year", "Temperature", "RH", "Ws"]] = df[["month", "day", "year", "Temperature", "RH", "Ws"]].astype(int)
```

```python
df.info()
```

```python
# Change the required columns into float data type
df[["Rain", "FFMC", "DMC", "DC", "ISI", "BUI", "FWI"]] = df[["Rain", "FFMC", "DMC", "DC", "ISI", "BUI", "FWI"]].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
# Lets save the cleaned dataset
df.to_csv("Cleanded_forest_fire_dataset.csv", index = False)
```

### EDA

```python
df_copy = df.drop(["day", "month", "year"], axis = 1)
```

```python
df_copy.head()
```

```python
# Categories in classes
df_copy["Classes"].value_counts()
```

```python
# Encoding of the categories in classes
df_copy["Classes"] = np.where(df["Classes"].str.contains("not fire"), 0, 1)
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
df_copy["Classes"].value_counts()
```

```python
# Plot density plot for all features
plt.style.use("seaborn-v0_8")
df_copy.hist(bins = 50, figsize = (20 ,15))
```

```python
# Percentage for Pie chart
percentage = df_copy["Classes"].value_counts(normalize = True) * 100
```

```python
# Plotting Pie Chart
classlabels = ["Fire", "Not Fire"]
plt.figure(figsize = (5, 10))
plt.pie(percentage, labels = classlabels, autopct = "%1.1f%%")
plt.title("Pie Chart of Classes")
```

```python
# Correlation
df_copy.corr()
```

```python
sns.heatmap(df_copy.corr())
```

```python
# Box Plot
sns.boxplot(df_copy["FWI"], color = "green") #type: ignore
```

```python
# Monthly Fire Analysis
dftemp = df.loc[df["Region"] == 0]
plt.subplots(figsize = (13, 5))
sns.set_style("whitegrid")
sns.countplot(x = "month", hue = "Classes", data = df)
plt.ylabel("No. of fires", weight = "bold")
plt.xlabel("Months", weight = "bold")
plt.title("Monthly Fire Analysis of Sidi Bel Region")
```
